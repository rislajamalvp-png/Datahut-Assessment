```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
```

```python
driver=webdriver.Chrome()

driver.get("https://www.adidas.co.in/men-shoes")

wait = WebDriverWait(driver, 20)
accept_button = wait.until(
    EC.element_to_be_clickable(
        (By.ID, "glass-gdpr-default-consent-accept-button")
    )
)
accept_button.click()
```

```python
Products=[]

cards=driver.find_elements(
    By.CSS_SELECTOR,
    "article[data-testid='plp-product-card']"
)
```

```python
products = []

start = 0

while True:

    url = f"https://www.adidas.co.in/men-shoes?start={start}"
    driver.get(url)

    cards = driver.find_elements(
        By.CSS_SELECTOR,
        "article[data-testid='plp-product-card']"
    )

    if len(cards) == 0:
        print("No more products found.")
        break

    print(f"Page {(start // 48) + 1} | Products: {len(cards)}")

    for card in cards:

        title = card.find_element(
            By.CSS_SELECTOR,
            "p[data-testid='product-card-title']"
        ).text

        product_link = card.find_element(
            By.CSS_SELECTOR,
            "a[data-testid='product-card-description-link']"
        ).get_attribute("href")

        sub_brand = card.find_element(
            By.CSS_SELECTOR,
            "p[data-testid='product-card-subtitle']"
        ).text

        main_price = card.find_element(
            By.CSS_SELECTOR,
            "div[data-testid='main-price']"
        ).text

        main_price = (
            main_price
            .replace("Price", "")
            .replace("Sale price", "")
            .strip()
        )

        original_price = card.find_elements(
            By.CSS_SELECTOR,
            "div[data-testid='original-price']"
        )

        if original_price:
            mrp = original_price[0].text.split("\n")[0]
            mrp = mrp.replace("Original price", "").strip()
        else:
            mrp = main_price

        products.append({
            "Title": title,
            "Product_link": product_link,
            "Sub_brand": sub_brand,
            "Sale_price": main_price,
            "MRP": mrp
        })

    start += 48

df = pd.DataFrame(products)

print(f"\nTotal Products Scraped : {len(df)}")
df
```

```python
df.to_csv(
    "adidas_products.csv",
    index=False,
    encoding="utf-8-sig"
)
```
