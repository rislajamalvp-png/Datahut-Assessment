```python
df["Sale_price"] = df["Sale_price"].str.replace("₹", "").str.replace(" ", "").str.replace(",", "").astype(float)
```

```python
df["MRP"]=df["MRP"].str.replace("₹","").str.replace(" ", "").str.replace(",", "").astype(float)
```

```python
df["Discount_Amount"]=df["MRP"]-df['Sale_price']
```

```python
df["Discount_pct"]=((df["Discount_Amount"]/df["MRP"])*100).round(2)
```
